## Project: Fisher's Tea Experiment

### Overview

A most beloved piece of [statistical lore](https://rss.onlinelibrary.wiley.com/doi/full/10.1111/j.1740-9713.2012.00620.x) about about the (most famous) statistician Ronald Fisher involves cups of tea with milk. Fisher and his friend and colleague, Dr. Muriel Bristol, worked at Cambridge in the 1920s and regularly had tea together. During one of their afternoon tea times, Bristol refused a cup of tea from Fisher because he put milk in first BEFORE pouring in the tea. Bristol said she could taste the difference, and much preferred the taste of tea when the milk was poured in afterward the tea. Fisher didn't think that there could be a difference and proposed a hypothesis test to examine the situation.

Fisher made 8 cups of tea, 4 with milk added in first and 4 with tea added in first, and gave them to Dr. Bristol without her seeing how they were made and she would say if she thought the tea or the milk was poured first. As it turned out, Bristol correctly identified if the tea or milk was poured first for all 8 of the cups. Fisher, being a skeptical statistician wanted to test if this could be happening by chance with Bristol just randomly guessing (or whether there was evidence against an assumption of Bristol just randomly guessing).

<summary style="color:blue"><u>Ethical Professionalism Considerations</u></summary>


Selecting and then specifying a good and accurate null hypothesis is an important skill to accuratly represent data and findings. Misrepresenting the null hypothesis can significantly mislead readers in ways such as:
    
1. Specificity of hypothesis. Null hypothesies often hide a lot of implicit information such as population distribution assumptions, any assumptions or estimate values you used to simplify the problem, etc.
    
2. Misleading readers by implying that a non-significant result means there is no effect, when in fact it may simply indicate insufficient evidence to reject the null hypothesis. This overlooks the possibility of sample issues, or Type II errors (which you will learn about later but essentially  meaning the test incorrectly concludes that there is no effect or difference when, in fact, there is one).
    
3. Drawing broad conclusions or making recommendations based on findings that reject the null hypothesis in a specific context. Once a null hypothesis is rejected, the alternative hypothesis is usually very general. Something like rejecting that there is no correlation between rain and pizza's delivered doesn't specify what this hypothetical correlation might be, we dont know if there are more pizzas delivered when it rains, or less, we can only theorize or conduct further experiements to continously reject more detailed hypothesies to narrow down the alternative hypothesis.
